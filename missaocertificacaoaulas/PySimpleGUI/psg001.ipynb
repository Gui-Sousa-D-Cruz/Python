```python
import PySimpleGUI as sg
sg.theme('Reddit')
#Criar as janelas e estilos(layout)
def janela_1():
    layout=[
        [sg.Text('Pronto para sua jornada, viajante?')],
        [sg.Push() ,sg.Button('Sim'), sg.Push(),sg.Button('Não'), sg.Push()]
    ]
    return sg.Window('RPG', layout=layout, finalize=True)

def janela_2():
    layout=[
        [sg.Text('Qual seu nome?')],
        [sg.Input(key='nome')],
        [sg.Push() ,sg.Button('Continuar'), sg.Push(), sg.Button('Voltar'), sg.Push()]
    ]
    return sg.Window('RPG', layout=layout, finalize=True)
    
nome = "" 
    
def janela_3():
    layout=[
        [sg.Text(f'Seja bem-vindo: {nome}!')],
        [sg.Push() ,sg.Button('Continuar'), sg.Push(), sg.Button('Voltar'), sg.Push()]
    ]
    
    return sg.Window('RPG', layout=layout, finalize=True)
#Criar a janelas
janela1 = janela_1()
janela2 = None
janela3 = None, None

#Criar um loop de leitura de eventos
while True:
    janela,evento,valor = sg.read_all_windows()
    
    #Fechar janela
    if janela == janela1 and evento == sg.WIN_CLOSED:
        break
    #Proxima janela
    if janela == janela1 and evento == 'Sim':
        janela2=janela_2()
        janela1.hide() 
        
        
    # Janela 2
    if janela == janela2 and evento == 'Continuar':
        nome = valor['nome']
        janela3=janela_3()
        janela2.hide()
        
    if janela == janela2 and evento == 'Voltar':
        janela2.hide()
        janela1.un_hide()
        
    # Janela 3
    if janela == janela3 and evento == 'Continuar':
        janela3.close()
        break
        
    if janela == janela2 and evento == 'Voltar':
        janela3.hide()
        janela2.un_hide()
        
        
   
        
    
    #Janela anterior
    
#Logica dos botões
```
